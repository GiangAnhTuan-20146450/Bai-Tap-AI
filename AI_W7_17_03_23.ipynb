```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np

url = 'mtcars.csv'
data = pd.read_csv(url)
data.head()
```

```python
data.hist()
```

```python
plt.show()
```

```python
sns.pairplot(data)
```

```python
correlation = data.corr()
```

```python
print(correlation)
```

```python
sns.heatmap(correlation,annot=True)
```

```python
data.plot(kind='scatter',x='wt',y='mpg',figsize=(9,9),color='black')
```

```python
from sklearn import linear_model
model = linear_model.LinearRegression()
model.fit(X=pd.DataFrame(data['wt']),y=data['mpg'])
print('Gia tri intercept:',model.intercept_)
print('Gia tri coeffient:',model.coef_)
print('Do chinh xac:',model.score(X=pd.DataFrame(data['wt']),y=data['mpg']))
```

```python
data.plot(kind='scatter',x='wt',y='mpg',figsize=(9,9),color='black',xlim=(0,7))
y_pred = model.predict(X=pd.DataFrame(data['wt']))
plt.plot(data['wt'],y_pred,color='blue')
```

```python
model=linear_model.LinearRegression()
X = pd.DataFrame([data['wt'],data['wt']**2]).T
model.fit(X,y=data['mpg'])
print('Intercept la:',model.intercept_)
print('Coefficient la:',model.coef_)
print('Do chinh xac:',model.score(X,y=data['mpg']))
```

```python
range = np.arange(1.5,5.5,0.1)
X_range=pd.DataFrame([range,range**2]).T
y_pred = model.predict(X=X_range)
data.plot(kind='scatter',x='wt',y='mpg',figsize=(9,9),color='black',xlim=(0,7))
plt.plot(range,y_pred,color='blue')
```

```python
model=linear_model.LinearRegression()
X = pd.DataFrame([data['wt'],data['wt']**2,data['wt']**3,data['wt']**4,data['wt']**5,data['wt']**6,data['wt']**7,data['wt']**8,data['wt']**9,data['wt']**10]).T
model.fit(X,y=data['mpg'])
print('Intercept la:',model.intercept_)
print('Coefficient la:',model.coef_)
print('Do chinh xac:',model.score(X,y=data['mpg']))

range = np.arange(1.5,5.5,0.1)
X_range=pd.DataFrame([range,range**2,range**3,range**4,range**5,range**6,range**7,range**8,range**9,range**10]).T
y_pred = model.predict(X=X_range)
data.plot(kind='scatter',x='wt',y='mpg',figsize=(9,9),color='black',xlim=(0,7))
plt.plot(range,y_pred,color='green')
```

```python
model = linear_model.LinearRegression()
model.fit(X=data.loc[:,['wt','hp']],y=data['mpg'])
print('Intercept la:',model.intercept_)
print('Coeffient la:',model.coef_)
print('Do chinh xac:',model.score(X=data.loc[:,['wt','hp']],y=data['mpg']))
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np

url = '50_Startups.csv'
data = pd.read_csv(url)
data.head()
```

```python
data.hist()
```

```python
correlation = data.corr()
```

```python
print(correlation)
```

```python
sns.heatmap(correlation,annot=True)
```

```python
data.plot(kind='scatter',x='R&D Spend',y='Profit',figsize=(9,9),color='black')
```

```python
from sklearn import linear_model
model = linear_model.LinearRegression()
model.fit(X=pd.DataFrame(data['R&D Spend']),y=data['Profit'])
print('Gia tri intercept:',model.intercept_)
print('Gia tri coeffient:',model.coef_)
print('Do chinh xac:',model.score(X=pd.DataFrame(data['R&D Spend']),y=data['Profit']))
```

```python
model=linear_model.LinearRegression()
X = pd.DataFrame([data['R&D Spend'],data['R&D Spend']**2,data['R&D Spend']**3,data['R&D Spend']**4,data['R&D Spend']**5,data['R&D Spend']**6,data['R&D Spend']**7,data['R&D Spend']**8,data['R&D Spend']**9,data['R&D Spend']**10]).T
model.fit(X,y=data['Profit'])
print('Intercept la:',model.intercept_)
print('Coefficient la:',model.coef_)
print('Do chinh xac:',model.score(X,y=data['Profit']))

range = np.arange(1.5,5.5,0.1)
X_range=pd.DataFrame([range,range**2,range**3,range**4,range**5,range**6,range**7,range**8,range**9,range**10]).T
y_pred = model.predict(X=X_range)
data.plot(kind='scatter',x='R&D Spend',y='Profit',figsize=(9,9),color='black',xlim=(0,150000))
plt.plot(range,y_pred,color='green')
```

```python
model = linear_model.LinearRegression()
model.fit(X=data.loc[:,['R&D Spend','Marketing Spend']],y=data['Profit'])
print('Intercept la:',model.intercept_)
print('Coeffient la:',model.coef_)
print('Do chinh xac:',model.score(X=data.loc[:,['R&D Spend','Marketing Spend']],y=data['Profit']))
```

```python
url = 'Advertising.csv'
data = pd.read_csv(url)
data.head()
```

```python
correlation = data.corr()
sns.heatmap(correlation,annot=True)
```

```python
from sklearn import linear_model
model = linear_model.LinearRegression()
model.fit(X=pd.DataFrame(data['TV']),y=data['sales'])
print('Gia tri intercept:',model.intercept_)
print('Gia tri coeffient:',model.coef_)
print('Do chinh xac:',model.score(X=pd.DataFrame(data['TV']),y=data['sales']))
```

```python
data.plot(kind='scatter',x='TV',y='sales',figsize=(9,9),color='black')
```

```python
data.plot(kind='scatter',x='TV',y='sales',figsize=(9,9),color='black',xlim=(0,350))
y_pred = model.predict(X=pd.DataFrame(data['TV']))
plt.plot(data['TV'],y_pred,color='blue')
```

```python
model=linear_model.LinearRegression()
X = pd.DataFrame([data['TV'],data['TV']**2]).T
model.fit(X,y=data['sales'])
print('Intercept la:',model.intercept_)
print('Coefficient la:',model.coef_)
print('Do chinh xac:',model.score(X,y=data['sales']))

range = np.arange(0,350,10)
X_range=pd.DataFrame([range,range**2]).T
y_pred = model.predict(X=X_range)
data.plot(kind='scatter',x='TV',y='sales',figsize=(9,9),color='black',xlim=(0,350))
plt.plot(range,y_pred,color='blue')
```
